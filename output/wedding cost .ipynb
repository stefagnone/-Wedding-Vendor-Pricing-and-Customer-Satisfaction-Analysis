```python
import pandas as pd

# Specify the file path for primary data analysis
file_path = "./__datasets/wedding_cost_data.xlsx"

# Read the CSV file into a pandas DataFrame
wedding = pd.read_excel(io=file_path)


def generate_department_table(department_data):
    department_table = f"\n{'='*40}\n{department_data['category'].iloc[0].capitalize()} Department\n{'='*40}\n"
    department_table += f"{'Item Name':<40}{'Price per Item':<20}{'Quantity':<10}{'Subtotal':<20}\n"
    department_table += '-'*90 + '\n'

    for i in range(len(department_data)):
        row = department_data.iloc[i]
        item_name = str(row['item_name'])
        price_per_item = str(row['price_per_item'])
        quantity = str(row['quantity'])
        subtotal = str(row['subtotal'])
        department_table += f"{item_name:<40}${price_per_item:<20}{quantity:<10}${subtotal:<20}\n"

    return department_table


def generate_wedding_cost_breakdown(data):
    budget_level = str(data['budget_level'].unique()[0].capitalize())
    total_cost = str(data['subtotal'].sum())

    wedding_cost_breakdown = f"{'='*50}\n{'Wedding Cost Breakdown':^50}\n{'='*50}\n"
    wedding_cost_breakdown += f"\nBudget Level: {budget_level}\n"
    wedding_cost_breakdown += f"\n{'Department-wise Cost Summary':^50}\n"

    departments = data['category'].unique()
    for department in departments:
        department_data = data[data['category'] == department]
        wedding_cost_breakdown += generate_department_table(department_data)

    wedding_cost_breakdown += f"\n{'Total Cost':<70}${total_cost}\n"
    return wedding_cost_breakdown

# Call the function with your DataFrame
cost_breakdown_string = generate_wedding_cost_breakdown(wedding)

# Print the generated string
print(cost_breakdown_string)
```
